```python
import numpy as np
```

```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("knee.jpg")
```

```python
plt.imshow(img)
```

```python
def sobelOperator(img):
    m = np.copy(img)
    size = m.shape
    for i in range(1, size[0] - 1):
        for j in range(1, size[1] - 1):
            sx = (img[i - 1][j - 1] + 2*img[i][j - 1] + img[i + 1][j - 1]) - (img[i - 1][j + 1] + 2*img[i][j + 1] + img[i + 1][j + 1])
            sy = (img[i - 1][j - 1] + 2*img[i - 1][j] + img[i - 1][j + 1]) - (img[i + 1][j - 1] + 2*img[i + 1][j] + img[i + 1][j + 1])
            m[i][j] = min(255, np.sqrt(sx**2 + sy**2))
    return m
    pass
```

```python
img = cv2.cvtColor(cv2.imread("knee.jpg"), cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
```

```python
img = sobelOperator(img)
```

```python
plt.imshow(img)
```
